```python
# Dated : 30 Oct 2019
# Making a basic line graph using Bokeh

# importing Bokeh
from bokeh.plotting import figure  # plotting is an interface of bokeh
from bokeh.io import output_file, show  # io is an interface of bokeh

# creating data --> var x and y of type list
x = [1,2,3,4,5]
# in bokeh number of values in each variable should be the same ie equal ie len(x) == len(y) is True
y = [6,7,8,9,10]
# creating output file in which graph will be shown using output_files,
# the file created " Line.html " is a webpage
output_file("Line.html")

# create figure object
f = figure() # f is like a box that holds the graph 

# creating the line plot
f.line(x,y)

# to show the graph, figure var is passed as arg.
show(f)
```

```python
# Dated : 30 Oct 2019
# Making a basic graph using Bokeh and pandas

# importing Bokeh and pandas
from bokeh.plotting import figure  # plotting is an interface of bokeh
from bokeh.io import output_file, show  # io is an interface of bokeh
import pandas as pd

# creating data --> var x and y of type list
df = pd.read_csv("data.csv")
x = df["x"]
# in bokeh number of values in each variable should be the same ie equal ie len(x) == len(y) is True
y = df["y"]
# creating output file in which graph will be shown using output_files,
# the file created " Line.html " is a webpage
output_file("Line graph using csv file.html")

# create figure object
f = figure() # f is like a box that holds the graph 

# creating the line plot
f.line(x,y)

# to show the graph, figure var is passed as arg.
show(f)
```

```python
# Dated : 30 Oct 2019
# Using data containing info about bachelors passed in USA
# using data file bachelors.csv
# creating graph for %age of women passed in data file

# importing bokeh and pandas
import pandas as pd
from bokeh.plotting import figure
from bokeh.io import output_file, show

# reading csv data file using pandas
df = pd.read_csv("bachelors.csv")

# creating variables for data 
x = df["Year"]
y = df["Engineering"]

# creating plot
f = figure(plot_width=500,plot_height=500,tools='pan')
f.line(x,y)
output_file("line graph for women in engg.html")
show(f)
```

```python
# Dated : 30 Oct 2019
# Using data containing info about weather in verlegenhuken
# using data file weather_data.csv
# creating graph for %age of women passed in data file

# importing bokeh and pandas
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show

# reading csv data file using pandas
df = pd.read_excel("weather_data.xlsx")

# creating variables for data 
df["Temperature"]=df["Temperature"]/10
df["Pressure"]=df["Pressure"]/10
 
p=figure(plot_width=1000,plot_height=800,tools='pan')
 
p.title.text="Temperature and Air Pressure"
p.title.text_color="Gray"
p.title.text_font="arial"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Temperature (Â°C)"
p.yaxis.axis_label="Pressure (hPa)"    
 
p.circle(df["Temperature"],df["Pressure"],size=0.5)
output_file("Weather.html")
show(p)
```
